## Split the data
- Split the data into independent variables and target variable
- Split the data into train and test set (80:20)

```python
train_len_standard = int(len(standard_df)*.8)
train_standard = standard_df[:train_len_standard]
test_standard = standard_df[train_len_standard:len(standard_df)]

train_len_robust = int(len(robust_df)*.8)
train_robust = robust_df[:train_len_robust]
test_robust = robust_df[train_len_robust:len(robust_df)]
```

```python
train_standard_features = train_standard.copy()
test_standard_features = test_standard.copy()

train_standard_labels = train_standard_features.pop('Close')
test_standard_labels = test_standard_features.pop('Close')
```

```python
train_robust_features = train_robust.copy()
test_robust_features = test_robust.copy()

train_robust_labels = train_robust_features.pop('Close')
test_robust_labels = test_robust_features.pop('Close')
```

## Modeling - FFNN

```python
def build_and_compile_model(input_dim=11):
    """
    Function to build and compile DNN architecture
    """
    model = keras.Sequential([
        layers.Dense(64, input_dim=input_dim, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])

    model.compile(loss='mean_squared_error',
                  optimizer=tf.keras.optimizers.Adam(0.001))
    return model

def fit_model(model, epochs, batch_size, train_features, train_labels):
    """
    Function to fit the DNN model with specified epochs and batch_size
    """
    print(model.summary())
    history = model.fit(train_features,
                        train_labels, validation_split=0.2,
                        verbose=0, epochs=epochs, batch_size=batch_size)
    return history

def plot_loss(history):
    """
    Function to plot history's loss
    """
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.xlabel('Epoch')
    plt.ylabel('Error [Close]')
    plt.legend()
    plt.grid(True)
    plt.show()
    
def generate_predictions(model, original_df, test_features, train_len):
    """
    Function to generate predictions with the developed model
    Returns predictions dataframe with Pred and Actual columns
    """
    predictions = model.predict(test_features).flatten()
    actual_close = original_df[['Close']]
    
    act = pd.DataFrame(actual_close.iloc[train_len:, 0])
    
    predictions = pd.DataFrame(predictions)
    predictions.reset_index(drop=True, inplace=True)
    predictions.index = test_features.index
    predictions['Actual'] = act['Close']
    predictions.rename(columns={0:'Pred'}, inplace=True)
    return predictions

def inverse_predictions(predictions_df, close_scaler, test_labels):
    """
    Function to inverse transform the predicted and actual values
    """
    inversed_pred = close_scaler.inverse_transform(np.array(predictions_df['Pred']).reshape(-1,1))
    inversed_act = close_scaler.inverse_transform(np.array(predictions_df['Actual']).reshape(-1,1))
    
    inversed = pd.DataFrame(inversed_pred)
    inversed['Actual'] = inversed_act
    inversed.rename({0:'Pred'}, axis=1, inplace=True)
    inversed.index = test_labels.index
    
    return inversed

def evaluate_model(inversed_df):
    rmse_ = rmse(inversed_df['Pred'], inversed_df['Actual'])
    mape_ = mape(inversed_df['Actual'], inversed_df['Pred'])
    rsquared_ = r2_score(inversed_df['Actual'], inversed_df['Pred'])
    return rmse_, mape_, rsquared_

def mape(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs((actual - pred) / actual)) * 100
```

### 1. Standard Scaler

```python
close_scaler_standard = load('../Feature Engineering/close_standard_scaler.bin')
```

```python
epochs_cands = [100,150,200]
batch_size_cands = [32,64,128]
standard_performances = {}
standard_models = {}
standard_predictions_dfs = {}
standard_inversed_dfs = {}
```

```python
modelID = 1
for epochs in epochs_cands:
    for batch_size in batch_size_cands:
        print(f'\n========== Model {modelID} with epochs: {epochs} and batch size: {batch_size} ==========')
        
        model = build_and_compile_model(17)
        history = fit_model(model, epochs, batch_size, train_standard_features, train_standard_labels)
        standard_models[f'Model-{modelID}'] = (model, history)
        
        predictions = generate_predictions(model, standard_df, test_standard_features, train_len_standard)
        standard_predictions_dfs[f'Model-{modelID}'] = predictions
        
        inversed = inverse_predictions(predictions, close_scaler_standard, test_standard_labels)
        standard_inversed_dfs[f'Model-{modelID}'] = inversed
        
        rmse_, mape_, rsquared_ = evaluate_model(inversed)
        standard_performances[f'Model-{modelID}'] = {
            'Epochs': epochs,
            'Batch Size': batch_size,
            'RMSE': rmse_,
            'MAPE': mape_,
            'R2': rsquared_
        }
        
        modelID += 1
```

```python
standard_performance_df = pd.DataFrame(standard_performances).transpose()
standard_performance_df[['Epochs', 'Batch Size']] = standard_performance_df[['Epochs', 'Batch Size']].astype(int)
standard_performance_df
```

Model-9 has the best performance with the lowest RMSE and MAPE.

#### Plot the best model

```python
def plot_predictions(predictions_df):
    """
    Function to plot predictions versus actual values
    """
    predictions_df['Actual'].plot(figsize=(20,8), legend=True, color='blue')
    predictions_df['Pred'].plot(legend=True, color='red', figsize=(20,8))
```
